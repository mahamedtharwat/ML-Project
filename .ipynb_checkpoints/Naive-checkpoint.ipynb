```python
import pandas as pd

data = pd.read_csv("Heart_Data_after_preprocessing.csv")

data.head(10)
```

```python
data.dtypes
```

```python
x = data.drop(columns=['HeartDisease'])
y = data['HeartDisease']


x.head(10)

```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x ,y , test_size=0.3,random_state=21)
```

```python
from sklearn.naive_bayes import GaussianNB

cat = GaussianNB()

```

```python
cat.fit(x_train,y_train)
```

```python
#cat.score(x_test , y_test)
print(f'the performance of training data : {round(cat.score(x_train , y_train),2)*100}%')
```

```python
y_pre=cat.predict(x_test)
```

```python
cat.predict_proba(x_test[:10])
```

```python
from sklearn.model_selection import cross_val_score

```

```python
accouracy_mean = cross_val_score(GaussianNB() , x_train , y_train , cv = 5)
```

```python
series_1 = pd.Series(accouracy_mean)

print(series_1.mean())

```

```python
from sklearn.metrics import accuracy_score
y_pred = cat.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy: {round((accuracy*100),2)}%')
```
