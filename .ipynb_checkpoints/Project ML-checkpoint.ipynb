```python
le = LabelEncoder()    # numeric data 
```

```python
df["HeartDisease"] = le.fit_transform(df["HeartDisease"])
df
```

```python
cols=["Smoking", "AlcoholDrinking", "Stroke", "PhysicalHealth", "MentalHealth", "DiffWalking", "Sex",
       "AgeCategory", "Diabetic", "PhysicalActivity", "SleepTime", "Asthma", "KidneyDisease","SkinCancer"]

for col in cols:
    df[col]=le.fit_transform(df[col])
```

```python
df = pd.get_dummies(df,drop_first=True) # Hot Encoding
df
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1:].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cl = KNeighborsClassifier(n_neighbors=100) 
cl.fit(x_train, y_train)
```

```python
y_pred = cl.predict(x_test)
```

```python
len(y_pred)
```

```python
from sklearn import metrics
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
```
