```python
cols=["HeartDisease","Smoking","AlcoholDrinking","Stroke","DiffWalking","Sex","AgeCategory","Diabetic","PhysicalActivity","Asthma","KidneyDisease","SkinCancer"]
```

```python
label_encoder = LabelEncoder()
for col in cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```
