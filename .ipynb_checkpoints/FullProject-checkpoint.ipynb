```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix , accuracy_score , recall_score
```

```python
df = pd.read_csv(r"C:\Users\Moham\OneDrive\Desktop\ProjectML\heart_2020.csv")
df
```

```python
df.info()
```

```python
cols=["HeartDisease" , "Smoking", "AlcoholDrinking", "Stroke", "PhysicalHealth", "MentalHealth", "DiffWalking", "Sex",
       "AgeCategory", "Diabetic", "PhysicalActivity", "SleepTime", "Asthma", "KidneyDisease","SkinCancer"]
le = LabelEncoder()
for col in cols:
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
df_dummies = pd.get_dummies(df ,columns=['GenHealth' , 'Race'] , drop_first=True , dtype=int)
df = df_dummies
```

```python
df
```

```python
x = df.drop(['HeartDisease'] , axis = 1)
y = df.HeartDisease
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.75,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = KNeighborsClassifier()
k.fit(x_train,y_train)
y_pred = k.predict(x_test)
```

```python
print(f'Accuracy : {round(accuracy_score(y_pred , y_test)*100 , 2)}') ## compare accuracy between y pred and y test
print(f'Recall : {round(recall_score(y_pred , y_test)*100 , 2)}') 
```

```python
count_class_0, count_class_1 = df.HeartDisease.value_counts()

dataset_class_0 = df[df["HeartDisease"] == 0]
dataset_class_1 = df[df["HeartDisease"] == 1]

dataset_class_1_over = dataset_class_1.sample(count_class_0, replace = True)
dataset_over = pd.concat([dataset_class_0, dataset_class_1_over], axis = 0)
print(dataset_over.HeartDisease.value_counts())
```

```python
print(f'Accuracy : {round(accuracy_score(y_pred , y_test)*100 , 2)}') ## compare accuracy between y pred and y test
print(f'Recall : {round(recall_score(y_pred , y_test)*100 , 2)}') 
```

```python
x_over = dataset_over.drop('HeartDisease' , axis = 1)
y_over = dataset_over.HeartDisease
```

```python
x_train_over , x_test_over , y_train_over , y_test_over = train_test_split(x_over,y_over,train_size=0.75,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = KNeighborsClassifier()
k.fit(x_train_over,y_train_over)
y_pred_over = k.predict(x_test_over)
```

```python
print(f'Accuracy : {round(accuracy_score(y_pred_over , y_test_over)*100 , 3)}')  # compare accuracy between y pred and y test
print(f'Recall : {round(recall_score(y_pred_over , y_test_over)*100 , 2)}') 
```

```python
from sklearn.ensemble import RandomForestClassifier
r = RandomForestClassifier(n_estimators=100)
r.fit(x_train_over,y_train_over)
y_pred_over = r.predict(x_test_over)
```

```python
print(f'Accuracy : {round(accuracy_score(y_pred_over , y_test_over)*100 , 3)}')  # compare accuracy between y pred and y test
print(f'Recall : {round(recall_score(y_pred_over , y_test_over)*100 , 2)}') 
```
